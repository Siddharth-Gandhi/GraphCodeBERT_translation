```python
import pandas as pd
import re
```

```python
# read csv file
df = pd.read_csv('exp.csv')

# Replace NaN in Class_Or_Function column with to False and Y to True
df['Class_Or_Function'] = df['Class_Or_Function'].fillna(False)
df['Class_Or_Function'] = df['Class_Or_Function'].replace('Y', True)
```

```python
# now write the java columns to a file called exp_java2cs.txt.java where each line contains the df['Java'][i] IN ONE LINE
# currently df['Java'][i] contains has \n in it, so you need to remove it and replace it with a space. Also it contains a bunch of formatting spaces, so you need to remove those too.
with open('exp_java2cs.txt.java', 'w') as f:
    for i in range(len(df['Java'])):
        code = df['Java'][i]
            # Remove single-line comments
        code = re.sub(r'//.*', '', code)

        # Remove multi-line comments
        code = re.sub(r'/\*.*?\*/', '', code, flags=re.DOTALL)

        # Replace multiple spaces with a single space
        code = re.sub(r'\s+', ' ', code)

        # Remove spaces before and after each line
        code = re.sub(r'^\s+|\s+?$', '', code)

        # Remove spaces before and after newlines
        code = re.sub(r'\s*\n\s*', '\n', code)
        f.write(code + '\n')
```

```python
# do the same for the C# column
with open('exp_java2cs.txt.cs', 'w') as f:
    for i in range(len(df['C#'])):
        code = df['C#'][i]
            # Remove single-line comments
        code = re.sub(r'//.*', '', code)

        # Remove multi-line comments
        code = re.sub(r'/\*.*?\*/', '', code, flags=re.DOTALL)

        # Replace multiple spaces with a single space
        code = re.sub(r'\s+', ' ', code)

        # Remove spaces before and after each line
        code = re.sub(r'^\s+|\s+?$', '', code)

        # Remove spaces before and after newlines
        code = re.sub(r'\s*\n\s*', '\n', code)
        f.write(code + '\n')
```
